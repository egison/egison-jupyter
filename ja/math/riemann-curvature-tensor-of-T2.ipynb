# еҹӢгӮҒиҫјгҒҝгғҲгғјгғ©гӮ№$T^2$гҒ®гғӘгғјгғһгғіжӣІзҺҮгғҶгғігӮҪгғ«

з®ЎеҚҠеҫ„$a$гҖҒеӨ§еҚҠеҫ„$b$гҒ®з’°зҠ¶гғҲгғјгғ©гӮ№гӮ’ж¬ЎгҒ®гӮҲгҒҶгҒ«еҹӢгӮҒиҫјгҒҝгҒҫгҒҷгҖӮ

$$
X(\theta,\phi)=\bigl((b+a\cos\theta)\cos\phi,
(b+a\cos\theta)\sin\phi,a\sin\theta\bigr),\qquad b>a>0.
$$

еҶ…еңЁзҡ„гҒ«е№іеқҰгҒӘе•Ҷз©әй–“гҒЁгҒ—гҒҰгҒ®гғҲгғјгғ©гӮ№гҒЁгҒҜз•°гҒӘгӮҠгҖҒгҒ“гҒ®еҹӢгӮҒиҫјгҒҝ
гғҲгғјгғ©гӮ№гҒ®жӣІзҺҮгҒҜе ҙжүҖгҒ«гӮҲгҒЈгҒҰз¬ҰеҸ·гҒҢеӨүгӮҸгӮҠгҒҫгҒҷгҖӮгҒ“гҒ“гҒ§гҒҜжҺҘз¶ҡгҖҒ
гғӘгғјгғһгғігғҶгғігӮҪгғ«гҖҒгғӘгғғгғҒгғҶгғігӮҪгғ«гҖҒгӮ№гӮ«гғ©гғјжӣІзҺҮгӮ’иЁҲз®—гҒ—гҒҫгҒҷгҖӮ


## еә§жЁҷгҒЁиЁҲйҮҸ

еә§жЁҷгғҷгӮҜгғҲгғ«гҒҜдә’гҒ„гҒ«зӣҙдәӨгҒ—гҖҒгҒқгҒ®й•·гҒ•гҒ®2д№—гҒҜ$a^2$гҒЁ
$(b+a\cos\theta)^2$гҒ§гҒҷгҖӮгҒ—гҒҹгҒҢгҒЈгҒҰгҖҒ

$$
g=\begin{pmatrix}a^2&0\\0&(b+a\cos\theta)^2\end{pmatrix}.
$$


```egison
declare symbol a, b, Оё, ПҶ: MathValue

def x : Vector MathValue := [| Оё, ПҶ |]
def X : Vector MathValue :=
  [| `(a * cos Оё + b) * cos ПҶ
   , `(a * cos Оё + b) * sin ПҶ
   , a * sin Оё
   |]

def e_i_j : Matrix MathValue := вҲӮ/вҲӮ X_j x~i

def g_i_j : Matrix MathValue :=
  generateTensor (\[u, v] -> V.* e_u_# e_v_#) [2, 2]

def g~i~j : Matrix MathValue := M.inverse g_#_#
```

## гғ¬гғҙгӮЈгғ»гғҒгғҙгӮЈгӮҝжҺҘз¶ҡ

ж¬ЎгҒ®ејҸгӮ’з”ЁгҒ„гҖҒз¬¬1ж·»еӯ—гӮ’$g^{ij}$гҒ§дёҠгҒ’гҒҫгҒҷгҖӮ
$\Gamma_{ijk}=\tfrac12(\partial_jg_{ik}+\partial_kg_{ij}-\partial_ig_{jk})$


```egison
def О“_i_j_k : Tensor MathValue :=
  (1 / 2) * (вҲӮ/вҲӮ g_i_k x~j + вҲӮ/вҲӮ g_i_j x~k - вҲӮ/вҲӮ g_j_k x~i)

def О“~i_j_k : Tensor MathValue := withSymbols [m]
  g~i~m . О“_m_j_k
```

## гғӘгғјгғһгғігғҶгғігӮҪгғ«

жң«е°ҫгҒ®2гҒӨгҒ®ж·»еӯ—гҒҜеҸҚеҜҫз§°гҒ§гҒҷгҖӮз¬¬1ж·»еӯ—гӮ’дёӢгҒ’гӮӢгҒЁгҖҒ$R_{ijkl}$гҒ§
гӮҲгҒҸзҹҘгӮүгӮҢгҒҹж·»еӯ—еҜҫгҒ®еҜҫз§°жҖ§гӮӮжҳҺгӮүгҒӢгҒ«гҒӘгӮҠгҒҫгҒҷгҖӮ


```egison
def R~i_j_k_l : Tensor MathValue := withSymbols [m]
  вҲӮ/вҲӮ О“~i_j_l x~k - вҲӮ/вҲӮ О“~i_j_k x~l
    + О“~m_j_l . О“~i_m_k - О“~m_j_k . О“~i_m_l

def R_i_j_k_l : Tensor MathValue := withSymbols [m]
  g_i_m . R~m_j_k_l
```

## гғӘгғғгғҒжӣІзҺҮгғ»гӮ№гӮ«гғ©гғјжӣІзҺҮгғ»гӮ¬гӮҰгӮ№жӣІзҺҮ

2ж¬Ўе…ғгҒ§гҒҜгҖҒгӮ№гӮ«гғ©гғјжӣІзҺҮгҒҜгӮ¬гӮҰгӮ№жӣІзҺҮгҒ®2еҖҚгҖҒгҒҷгҒӘгӮҸгҒЎ
$\mathcal R=2K$гҒ§гҒҷгҖӮгҒ“гҒ®гғҲгғјгғ©гӮ№гҒ§гҒҜгҖҒ

$$
K(\theta)=\frac{\cos\theta}{a(b+a\cos\theta)}.
$$


```egison
def Ric_i_j : Matrix MathValue := withSymbols [m]
  sum (contract R~m_i_m_j)

def scalarCurvature : MathValue := withSymbols [i, j]
  g~i~j . Ric_i_j

def gaussianCurvature : MathValue := scalarCurvature / 2
```

## е№ҫдҪ•еӯҰзҡ„гҒӘж„Ҹе‘і

гғҲгғјгғ©гӮ№гҒ®еӨ–еҒҙпјҲ$\cos\theta>0$пјүгҒ§гҒҜжӣІзҺҮгҒҢжӯЈгҖҒеҶ…еҒҙ
пјҲ$\cos\theta<0$пјүгҒ§гҒҜиІ гҒЁгҒӘгӮҠгҖҒдёҠз«ҜгҒЁдёӢз«ҜгҒ®еҶҶе‘ЁдёҠгҒ§гҒҜжӣІзҺҮгҒҢ0гҒ§
ж”ҫзү©зҡ„гҒ«гҒӘгӮҠгҒҫгҒҷгҖӮ
гӮ¬гӮҰгӮ№вҖ“гғңгғігғҚз©ҚеҲҶгҒ§гҒҜжӯЈиІ гҒ®еҜ„дёҺгҒҢжү“гҒЎж¶ҲгҒ—еҗҲгҒ„гҖҒ
$\chi(T^2)=0$гҒЁж•ҙеҗҲгҒ—гҒҫгҒҷгҖӮ
