# Demo 2: Hodge Laplacian in Polar Coordinates

This demo computes the Hodge Laplacian of a scalar function $f(r,\theta)$.
The key definitions are ordinary tensor functions:

$$
dA = \frac{\partial A}{\partial x^i} dx^i,
\qquad
\delta A = (-1)^{N(k+1)+1}\ast d\ast A,
\qquad
\Delta A = \delta d A
$$

for a $0$-form $A$ in two dimensions.


## Polar metric

The coordinates are $x=(r,\theta)$ and the metric is

$$
g =
\begin{pmatrix}
1 & 0\\
0 & r^2
\end{pmatrix}.
$$


```egison
declare symbol r, θ: MathValue

def N : Integer := 2

def x : Vector MathValue := [| r, θ |]

def g_i_j : Matrix MathValue := [| [| 1, 0 |], [| 0, r^2 |] |]_i_j
def g~i~j : Matrix MathValue := [| [| 1, 0 |], [| 0, r^(-2) |] |]~i~j
```

```egison
g_#_#
```

## Exterior derivative and Hodge star


```egison
def d (A: Tensor MathValue) : Tensor MathValue := !(flip ∂/∂) x A

def hodge (A: Tensor MathValue) : Tensor MathValue :=
  let k := dfOrder A in
    withSymbols [i, j]
      (sqrt (M.det g_#_#)) *
        (foldl (.)
          ((subrefs A (map 1#j_$1 (between 1 k))) .
           (subrefs (ε' N k) (map 1#i_$1 (between 1 N))))
          (map 1#g~(i_$1)~(j_$1) [1..k]))
```

## Codifferential and Hodge Laplacian


```egison
def δ (A: Tensor MathValue) : Tensor MathValue :=
  let k := dfOrder A in
    -1^(N * (k + 1) + 1) * (hodge (d (hodge A)))

def Δ (A: Tensor MathValue) : Tensor MathValue :=
  match (dfOrder A) as integer with
  | #0 -> δ (d A)
  | #N -> d (δ A)
  | _  -> d (δ A) + δ (d A)

def f : MathValue := function (r, θ)
```

## Result

The output is the polar-coordinate Hodge Laplacian:

$$
\Delta f
= -\frac{\partial^2 f}{\partial r^2}
  - r^{-1}\frac{\partial f}{\partial r}
  - r^{-2}\frac{\partial^2 f}{\partial \theta^2}.
$$
