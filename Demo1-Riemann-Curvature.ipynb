# Demo 1: Riemann Curvature on $S^2$

This demo computes the Riemann curvature tensor of the 2-sphere in two ways.

1. Directly from the Riemann curvature tensor formula

$$
R^i{}_{jkl}
  = \frac{\partial \Gamma^i{}_{jl}}{\partial x^k}
  - \frac{\partial \Gamma^i{}_{jk}}{\partial x^l}
  + \Gamma^m{}_{jl}\Gamma^i{}_{mk}
  - \Gamma^m{}_{jk}\Gamma^i{}_{ml}.
$$

2. From the curvature-form formula

$$
\Omega^i{}_{j} = d\omega^i{}_{j} + \omega^i{}_{k}\wedge\omega^k{}_{j}.
$$


## Metric on $S^2$

The coordinates are $x=(\theta,\phi)$.  We embed the sphere in
$\mathbb{R}^3$, differentiate its position vector to obtain the coordinate
tangent vectors, and take their dot products.  The resulting metric is

$$
g =
\begin{pmatrix}
r^2 & 0\\
0 & r^2\sin^2\theta
\end{pmatrix}.
$$


```egison
declare symbol r, őł, ŌÜ: MathValue

def x : Vector MathValue := [| őł, ŌÜ |]

def X : Vector MathValue :=
  [| r * sin őł * cos ŌÜ, r * sin őł * sin ŌÜ, r * cos őł |]

def e_i_j : Matrix MathValue := ‚ąā/‚ąā X_j x~i

def g_i_j : Matrix MathValue :=
  generateTensor (\[a, b] -> V.* e_a_# e_b_#) [2, 2]

def g~i~j : Matrix MathValue := M.inverse g_#_#
```

```egison
g_#_#
```

```egison
g~#~#
```

## Christoffel symbols

The first-kind Christoffel symbols are defined directly from the metric, and the
second-kind symbols are obtained by raising one index.


```egison
def őď_i_j_k : Tensor MathValue :=
  (1 / 2) * (‚ąā/‚ąā g_i_k x~j + ‚ąā/‚ąā g_i_j x~k - ‚ąā/‚ąā g_j_k x~i)

def őď~i_j_k : Tensor MathValue := withSymbols [m]
  g~i~m . őď_m_j_k
```

```egison
őď~1_#_#
```

```egison
őď~2_#_#
```

## Path A: direct Riemann curvature tensor

Now we translate the tensor formula for $R^i{}_{jkl}$ directly into Egison.


```egison
def R~i_j_k_l : Tensor MathValue := withSymbols [m]
  ‚ąā/‚ąā őď~i_j_l x~k - ‚ąā/‚ąā őď~i_j_k x~l + őď~m_j_l . őď~i_m_k - őď~m_j_k . őď~i_m_l
```

```egison
R~#_#_1_2
```

```egison
R~#_#_2_1
```

## Path B: curvature forms

The exterior derivative and wedge product reuse the same tensor machinery. The
connection form is $\omega^i{}_j = \Gamma^i{}_{jk}\,dx^k$.


```egison
def d (t : Tensor MathValue) : Tensor MathValue := !(flip ‚ąā/‚ąā) x t

def ŌČ~i_j : Matrix MathValue := őď~i_j_#

def ő©~i_j : Tensor MathValue := withSymbols [k]
  antisymmetrize (d ŌČ~i_j + ŌČ~i_k ‚ąß ŌČ~k_j)
```

```egison
ő©~#_#_1_2
```

```egison
ő©~#_#_2_1
```

The curvature-form components contain the standard $\frac12$ normalization for
antisymmetric 2-forms, but they encode the same curvature data as the direct
Riemann tensor calculation.
